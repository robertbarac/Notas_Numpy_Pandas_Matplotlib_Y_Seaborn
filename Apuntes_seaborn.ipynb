# 14 - Categóricos
¿Cómo manejar variables categóricas?

```python
#Conteo de días que se dejan propinas por día, diferenciado por sexo
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
#
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color="0", marker='<')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='pastel')
plt.show()
```

```python
sns.violinplot(data=tips, y='total_bill')
plt.show()
```

```python
sns.violinplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) #grafica en strippplot por defecto
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='swarm')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='box')
plt.show()
```

```python
# Mostrar cómo dan propinas por día, hora de comida(Almurezo o cena) y según el sexo
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='box', col='time')
plt.show()
```

# 16 - Jointplot y Pairplot

```python
#Es un scatter combinado con distribuciones de barra en x e y.
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True)
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True,
              marginal_kws=dict(bins=25, fill=False, multiple='dodge'))
plt.show()
```

**Pairplot()**

```python
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
plt.show()
```

# 17 - Heatmap

```python
#correlación entre las variables numéricas del dataset
tips.corr()
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True)
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black')
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black', vmin=0.5, vmax=1, cbar=True)
plt.show()
```
