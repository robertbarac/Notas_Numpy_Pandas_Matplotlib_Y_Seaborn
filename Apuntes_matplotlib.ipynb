#Subplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# Primer nÃºmero: filas, segundo nÃºmero: columnas, tercer nÃºmero: Ã­ndice del grÃ¡fico en cuestiÃ³n
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')

plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'b:')

plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

#MÃ©todo orientado a objetos

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
# los dos primeros valores/parÃ¡metros corresponden a la posiciÃ³n de inicion de la lÃ­nea
# los otros dos parÃ¡metros es las dimensiones del grÃ¡fico
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x, y, 'b')
fig.show()
```

```python
fig = plt.figure()
# los dos primeros valores/parÃ¡metros corresponden a la posiciÃ³n de inicion de la lÃ­nea
# los otros dos parÃ¡metros es las dimensiones del grÃ¡fico
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
axes2.set_facecolor('cyan')
fig.show()
```

# Subplots

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y), 'b'
```

```python
# nrows ws nÃºmero de filas
# ncols es nÃºmero de columnas, fÃ¡cil
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'b')
ax2.plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x), 'y')
axes[0,2].plot(x, np.tan(x), 'r')
axes[0,3].plot(x, np.tanh(x), 'g')
```

```python
fig, ((ax1 , ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.tanh(x), 'g')
ax5.plot(x, x**2)
ax6.plot(x, 1/x, 'y')
ax7.plot(x, x**(1/2), 'r')
ax8.plot(x, np.sinh(x), 'g')
fig.tight_layout()
```

# Leyendas, etiquetas, tÃ­tulos, tamaÃ±o.

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y)
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y, 'r', label="$sin(x)$")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
#lo contrario de OO...
plt.plot(x, y, label= 'sin(x)')
plt.title('este es un titulo')
plt.xlabel='x'
plt.ylabel='y'
plt.legend(loc='lower left')
#plt.legend(loc='lower right')
#plt.legend(loc='upper left')
#plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label= 'sin(x)')
plt.title('este es un titulo')
plt.xlabel='x'
plt.ylabel='y'
plt.legend(loc='lower left', bbox_to_anchor=(1, 0.5))
#plt.legend(loc='lower right')
#plt.legend(loc='upper left')
#plt.legend(loc='lower right')
```

# BarPlot

```python
x = country = ['india', 'japan', 'mexico', 'colombia', 'germany']
population = [1000, 200, 160, 50, 40]
```

```python
plt.bar(country, population, width=0.5, color=['red', 'blue']) # width es para el ancho de la barra
# color es para alternar visualmente entre el color de la barra del paÃ­s.
plt.xticks(np.arange(5),('India', 'JapÃ³n', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.yticks(rotation=45)
plt.title('Robert')
plt.xlabel = 'PaÃ­s'
plt.ylabel = 'Poblacion'
plt.show()
```

```python
plt.barh(country, population) # width es para el ancho de la barra
# color es para alternar visualmente entre el color de la barra del paÃ­s.
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.title('Robert')
plt.xlabel = 'PaÃ­s'
plt.ylabel = 'Poblacion'
plt.show()
```
