#19 - Combinando DataFrames
*Inner join* es s√≥lo los registros que tengan columndas en com√∫n, en SQL ser√≠an donde la llave primaria de una tabla coincida con la llave for√°nea de la otra tabla donde haya una relaci√≥n.
 *left join* hace lo de inner con el agregado de los datos de la primera tabla (del dataframe en este caso) que no tengan una relaci√≥n con la otra tabla o dataframe.
 *rigth join* ser√≠a con la segunda tabla √≥ dataframe
 *outer join* juntar√≠a todo, tengan relaci√≥n o no.

**Merge**

```python
izq = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
 'A':['A0', 'A1', 'A2', 'A3'],
 'B':['B0', 'B1', 'B2', 'B3']
 })

der = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
 'C':['C0', 'C1', 'C2', 'C3'],
 'D':['D0', 'D1', 'D2', 'D3']
 })
```

```python
#Mezcla de dataframes
izq.merge(der, on='key') #onkey es para indicar la 'llave', la columna que tienen en com√∫n las tablas...
```

**¬øQu√© pasa si la columna por la cual queremos fusionar los dataframes tienen nombres diferentes aunque tengan los mismos datos?**

```python
izq = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
 'A':['A0', 'A1', 'A2', 'A3'],
 'B':['B0', 'B1', 'B2', 'B3']
 })

der = pd.DataFrame({'key_2':['k0', 'k1', 'k2', 'k3'],
 'C':['C0', 'C1', 'C2', 'C3'],
 'D':['D0', 'D1', 'D2', 'D3']
 })
```

```python
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
 'A':['A0', 'A1', 'A2', 'A3'],
 'B':['B0', 'B1', 'B2', 'B3']
 })

der = pd.DataFrame({'key_2':['k0', 'k1', 'k2', np.nan],
 'C':['C0', 'C1', 'C2', 'C3'],
 'D':['D0', 'D1', 'D2', 'D3']
 })
```

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='left') # prioridad para los datos de la izquierda que no tienen match con los de la derecha
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right') # prioridad para los datos de la derecha que no tienen match con los de la izquierda
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='inner')
```

**Melt**

El m√©todo melt toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas para especificar la antigua columna y el valor que tra√≠a.

Por ejemplo, simplemente al imprimir las cinco primeras filas del DataFrame con las columnas de Name y Genre se tiene este resultado.

Para ello ejecuta la siguiente l√≠nea en tu Jupyter Notebook:

```python
df_books[['Name','Genre']].head(5)
```

```python
df_books[['Name','Genre']].head(5).melt()
```

Ahora cada resultado de las dos columnas pasa a una fila de este modo a tipo llave:valor.

En el siguiente ejemplo ejecutemos melt de esta manera:

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```

Simplemente, podemos seleccionar las columnas que no quiero hacer melt usando el par√°metro id_vars. Para este caso Year y tambi√©n la √∫nica columna que quiero aplicar el melt, para este caso Genre con la propiedad value_vars.
